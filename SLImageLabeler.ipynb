```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
def opener(path,scalefac,savename):
    fig = Image.open(path)
    arr = np.flip(np.transpose(np.array(fig)[:,:,0]),axis=1)
    plt.figure(figsize=(14,14))
    plt.imshow(arr,cmap='gray',origin='lower',vmin=0,vmax=np.max(arr)/scalefac)
    plt.xlabel('Horizontal Pixel #',{'fontsize':'15'})
    plt.ylabel('Vertical Pixel #',{'fontsize':'15'})
    plt.savefig(savename,bbox='tight')
    return arr;
```
