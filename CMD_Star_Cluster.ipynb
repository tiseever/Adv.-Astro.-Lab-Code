```python
def cs(x,y):
    Matched_Array = np.zeros(Absolute_MV.shape)
    for i in range(len(Absolute_MV)):
        Flag = 0
        for j in range(len(y.values)):
            if(Flag==1):
                continue
            elif(np.abs(x.values[j]-Absolute_MV[i])<1):
                Matched_Array[i] = y.values[j]
                Flag = 1
            else:
                continue
    cs = sp.chisquare(Absolute_MV,Matched_Array)
    return cs
```

```python
CS1 = cs(Vmag1,ColorI1)[0]/len(Absolute_MV)
CS2 = cs(Vmag2,ColorI2)[0]/len(Absolute_MV)
CS3 = cs(Vmag3,ColorI3)[0]/len(Absolute_MV)
CS4 = cs(Vmag4,ColorI4)[0]/len(Absolute_MV)
CS5 = cs(Vmag5,ColorI5)[0]/len(Absolute_MV)
```
